```python

df = spark.read.format("csv") \
    .options(**configs) \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(bronze_path)

```

```python
df.show(10)
```

```python
%sql
CREATE CATALOG IF NOT EXISTS nifty_catalog;
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS nifty_catalog.gold_schema;
```

```python
df.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("nifty_catalog.nifty_schema.stock_raw")
```
