```python
df_bronze = spark.read.table("nifty_catalog.nifty_schema.stock_raw")
```

```python

df_silver = spark.read.format("csv") \
    .options(**configs) \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(df_bronze)
w
```

```python
df_silver.show(10)
```

```python
from pyspark.sql.functions import col
from pyspark.sql.types import StringType, DateType, DoubleType, LongType

# Fix schema
df_silver = df_silver \
    .withColumn("symbol", col("symbol").cast(StringType())) \
    .withColumn("date", col("date").cast(DateType())) \
    .withColumn("open", col("open").cast(DoubleType())) \
    .withColumn("high", col("high").cast(DoubleType())) \
    .withColumn("low", col("low").cast(DoubleType())) \
    .withColumn("close", col("close").cast(DoubleType())) \
    .withColumn("volume", col("volume").cast(LongType()))

# Show corrected schema
df_silver.printSchema()

# Show data
df_silver.show(5)
```

```python
df_silver = df_silver.dropDuplicates()
```

```python
df_silver = df_silver.dropna()
```

```python
df_silver.show(10)
```

```python
df_silver.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("nifty_catalog.nifty_schema.stock_cleaned")
```
