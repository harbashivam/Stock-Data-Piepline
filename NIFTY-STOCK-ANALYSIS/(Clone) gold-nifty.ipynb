```python
df = spark.table("nifty_catalog.nifty_schema.stock_cleaned")
```

```python
from pyspark.sql.functions import avg, max, min, sum

df_gold = df.groupBy("symbol") \
    .agg(
        avg("close").alias("avg_close_price"),
        max("high").alias("max_price"),
        min("low").alias("min_price"),
        sum("volume").alias("total_volume")
    )
```

```python
df_gold.show(10)
```

```python
df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("nifty_catalog.nifty_schema.gold_nifty")
```

```python
%sql
CREATE OR REPLACE TABLE nifty_catalog.nifty_schema.gold_daily AS
SELECT
    symbol,
    date AS day,
    AVG(close) AS avg_close_price,
    MAX(high) AS max_price,
    MIN(low) AS min_price,
    SUM(volume) AS total_volume
FROM nifty_catalog.nifty_schema.gold_nifty
WHERE symbol = '${symbol_param}'
GROUP BY symbol, date
ORDER BY day;
```

```python
%sql
CREATE OR REPLACE TABLE nifty_catalog.nifty_schema.gold_daily AS
SELECT
    symbol,
    date AS day,
    FIRST(open) AS open_price,
    LAST(close) AS close_price,
    AVG(close) AS avg_close_price,
    MAX(high) AS max_price,
    MIN(low) AS min_price,
    SUM(volume) AS total_volume
FROM nifty_catalog.nifty_schema.gold_nifty
WHERE symbol = '${symbol_param}'
GROUP BY symbol, date
ORDER BY day;
```

```python
%sql
--
SELECT
    symbol,
    DATE_TRUNC('month', date) AS month,
    AVG(close) AS avg_close_price,
    MAX(high) AS max_price,
    MIN(low) AS min_price,
    SUM(volume) AS total_volume
FROM nifty_catalog.nifty_schema.gold_nifty
WHERE symbol = '${symbol_param}'
GROUP BY symbol, DATE_TRUNC('month', date)
ORDER BY month;
```

```python
%sql
CREATE OR REPLACE TABLE nifty_catalog.nifty_schema.gold_monthly AS
SELECT
    symbol,
    DATE_TRUNC('month', date) AS month,
    FIRST(open) AS open_price,
    LAST(close) AS close_price,
    AVG(close) AS avg_close_price,
    MAX(high) AS max_price,
    MIN(low) AS min_price,
    SUM(volume) AS total_volume
FROM nifty_catalog.nifty_schema.gold_nifty
WHERE symbol = '${symbol_param}'
GROUP BY symbol, DATE_TRUNC('month', date)
ORDER BY month;
```
